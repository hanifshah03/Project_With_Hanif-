```python
import pandas as pd
import os
from sklearn.ensemble import RandomForestClassifier

# 1. Automatically locate the exact folder path for the Titanic files
dataset_dir = ""
for dirname, _, filenames in os.walk('/kaggle/input'):
    if 'train.csv' in filenames:
        dataset_dir = dirname
        break

print(f"Found the data files inside: {dataset_dir}")

# 2. Load the files dynamically using the found path
train_df = pd.read_csv(os.path.join(dataset_dir, "train.csv"))
test_df = pd.read_csv(os.path.join(dataset_dir, "test.csv"))

# 3. Extract basic features for training
features = ["Pclass", "Sex", "SibSp", "Parch"]
X = pd.get_dummies(train_df[features])
y = train_df["Survived"]
X_test = pd.get_dummies(test_df[features])

# 4. Train a Random Forest ML model
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)

# 5. Generate the predictions
predictions = model.predict(X_test)

# 6. Format the final output file for Kaggle
output = pd.DataFrame({'PassengerId': test_df.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)

print("--- SUCCESS ---")
print("Day 1 Challenge Code Executed! 'submission.csv' generated successfully.")
```
